# filtering with some conditions on the datasets

```python
dat = {
    'Name': ['Alice', 'Bob', 'Charli', 'David', 'Emily'],
    'Age': [20, 30, 35, 40, 45],
    'Gender': ['F', 'M', 'M', 'M', 'F']
}

dt1 = pd.DataFrame(dat)
dt1
```

```python
dt1[dt1['Age']>30]

dt1[(dt1['Age']>30) & (dt1['Gender']=='M')]

dt1[dt1['Name'].isin(['Emily'])]

dt1[dt1['Name'].str.startswith('E')]

dt1.sort_values(by='Age', ascending=False)

dt1['Marks']=[90, 80, 75, 95, 89]
dt1

dt1.drop(columns='Marks')

dt1.rename(columns={'Name':'Names'})

dt1.groupby("Gender")

dt1.groupby("Gender")["Age"].mean()
```
